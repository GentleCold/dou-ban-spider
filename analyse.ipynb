```python
import pandas as pd
with open('data.csv', 'r', encoding='utf_8_sig') as f:
    data = pd.read_csv(f)
data
```

出版年份统计

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False

x = data.groupby('6')['6'].count().keys()
y = data.groupby('6')['6'].count()
plt.figure(figsize=(50, 10))
plt.bar(range(len(x)), y, tick_label=x)
plt.show()
```

评分统计

```python
x = data.groupby('8')['8'].count()[1:].keys()
y = data.groupby('8')['8'].count()[1:]
plt.figure(figsize=(50, 10))
plt.bar(range(len(x)), y, tick_label=x)
plt.show()
```

评分与评论数散点图

```python
x = data.loc[data['8']!=0.0, '8']
y = data.loc[data['8']!=0.0, '9']
plt.figure(figsize=(10, 10))
plt.scatter(x, y)
plt.show()
```

评分与年份散点图

```python
x = data.loc[data['8']!=0.0, '8']
y = data.loc[data['8']!=0.0, '6']
plt.figure(figsize=(10, 10))
plt.scatter(x, y)
plt.show()
```

类型分布

```python
x = data.groupby('3')['3'].count().keys()
y = data.groupby('3')['3'].count()
plt.figure(figsize=(10, 10))
plt.pie(y, labels=x, autopct='%1.2f%%')
plt.show()
```
